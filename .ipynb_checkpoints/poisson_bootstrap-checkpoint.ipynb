## Poisson Bootstrap example

### Normals with mean in 100 and 95 with 10000 elements

```python
a = np.random.normal(100, 5, 10000)
```

```python
fig = plt.figure(figsize=(8, 2))
plt.hist(a, bins=75, color=saturate_color(palette[4], 0.75))
plt.xlim(70, 120)
axis_rstyle()
plt.show()
savefig('a', dir=dir_save_img)
```

```python
b = np.random.normal(95, 5, 10000)
```

```python
fig = plt.figure(figsize=(8, 2))
plt.hist(b, bins=75, color=saturate_color(palette[5], 0.75));
plt.xlim(70, 120)
axis_rstyle()
plt.show()
savefig('b', dir=dir_save_img)
```

```python
test_poisson_bootstrap(
    data1=a,
    data2=b,
    colors=[palette[4], palette[5]])
```
