## Poisson Bootstrap example

### Normals with mean in 100 and 95 with 10000 elements

```python
a = np.random.normal(100, 5, 1000)
```

```python
fig = plt.figure(figsize=(8, 2))
plt.hist(a, bins=60, color=alpha_color(palette[1], 0.85))
plt.xlim(70, 120)
axis_rstyle()
plt.show()
savefig('a', dir=dir_save_img)
```

```python
b = np.random.normal(95, 5, 1000)
```

```python
fig = plt.figure(figsize=(8, 2))
plt.hist(b, bins=60, color=alpha_color(palette[2], 0.8));
plt.xlim(70, 120)
plt.ylim(0, 60)
axis_rstyle()
plt.show()
savefig('b', dir=dir_save_img)
```

```python
test_poisson_bootstrap(
    data1=a,
    data2=b,
    colors=[palette[1], palette[2]])
```
